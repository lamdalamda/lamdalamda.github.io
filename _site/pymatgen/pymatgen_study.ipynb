pymatgen: 通过core.structure 构建结构、晶体
原子在pymatgen.core.Element目录下，例如pymatgen.core.Element.Si
可以输入string也可以打进去Element.Si对象

```python
import pymatgen
from pymatgen.core.structure import Molecule
from pymatgen.core.structure import Structure
#分子
from pymatgen.core import Element

sio=Molecule(["Si","O"],[[0,0,0],[0.5,0.5,0.5]])

#生成lattice

fccsi=Structure.from_spacegroup("Fm-3m",pymatgen.core.Lattice.cubic(3.9),["Si"],[[0,0,0]])
fccsi=Structure.from_spacegroup("Fm-3m",pymatgen.core.Lattice.cubic(3.9),[Element.Si],[[0,0,0]])
# 通过这个lattice生成10*20*5的supercell

fccsi_supercell=fccsi*(10,20,5)
print(fccsi_supercell)
```

pymatgen.io vasp输出

```python
#通过上面建立的structure对象来生成POSCAR
from pymatgen.io import vasp

siposcar=vasp.Poscar(fccsi)
siposcar.write_file("POSCAR_matgen")
```

pymatgen.io vasp输入
通过 vasp.Poscar 对象的from_file 函数读取POSCAR文件
注意POSCAR init时候需要随便塞进去一个structure

```python
from_Ni_100=vasp.Poscar(fccsi).from_file("POSCAR")
print(from_Ni_100.get_string())
from_Ni_100.write_file("POSCAR.Ni_100")
```
